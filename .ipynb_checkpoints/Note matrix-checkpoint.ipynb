```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('fragrancedata.csv')
df = df.drop(columns={'Unnamed: 0'})
```

```python
df = df.drop_duplicates()
```

# Base notes

```python
base_notes = []
for i in df['Начальная нота']:
    try:
        for j in i.split(','):
            note = j.strip()
            base_notes.append(note) if note not in base_notes else base_notes
    except:
        continue
base_notes.sort()
```

# Heart notes

```python
heart_notes = []
for i in df['Нота сердца']:
    try:
        for j in i.split(','):
            note = j.strip()
            heart_notes.append(note) if note not in heart_notes else heart_notes
    except:
        continue
heart_notes.sort()
```

# Top notes

```python
top_notes = []
for i in df['Конечная нота']:
    try:
        for j in i.split(','):
            note = j.strip()
            top_notes.append(note) if note not in top_notes else top_notes
    except:
        continue
top_notes.sort()
```

```python
notes = []
notes = base_notes + heart_notes + top_notes
notes = list(set(notes))
notes.sort()
```

```python
df_new = pd.DataFrame(columns=notes)
```

```python
df_new['Perfume'] = df_notes['Perfume']
```

```python
for i in range(len(df_notes.head(10))):
    for j in df_notes['Начальная нота'].values[i].split(','):
        if not df_new[j.strip()].empty:
            df_new[j.strip()].values[i] = 1
    for j in df_notes['Нота сердца'].values[i].split(','):
        if not df_new[j.strip()].empty:
            df_new[j.strip()].values[i] = 1
    for j in df_notes['Конечная нота'].values[i].split(','):
        if not df_new[j.strip()].empty:
            df_new[j.strip()].values[i] = 1
df_new.fillna(0,inplace=True)
```
