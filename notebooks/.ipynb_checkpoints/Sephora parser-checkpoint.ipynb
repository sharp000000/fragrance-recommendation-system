# Parsing information about exact parfume

```python
parsed_html_one = BeautifulSoup(scraper.get("https://makeup.com.ua/product/{}/".format(127021)).content,'html.parser')
```

```python
columns = []
for length in parsed_html_one.find_all("strong"):
    columns.append(length.text[:-1])
columns = columns[:9]
columns.append('Code')
columns.append('Brand')
columns.append('Perfume')
fragrance = pd.DataFrame(columns=columns)
```

```python
def addParfume(page):
    parsed_html_one = BeautifulSoup(scraper.get("https://makeup.com.ua/product/{}/".format(page)).content,'html.parser')
    n = len(fragrance)+1
    for col in parsed_html_one.find_all("strong")[:9]:
        try:
            fragrance.loc[n,col.text[:-1]] = col.next_sibling.strip()
        except:
            continue
    fragrance.loc[n,'Code'] = page
    fragrance.loc[n,'Brand'] = parsed_html_one.find_all('span',itemprop="name")[3].text
    fragrance.loc[n,'Perfume'] = parsed_html_one.find_all('span',itemprop="name")[2].text
```

```python
for i in range(1,pages):
    print('Page: ',i, 36*i)
    parsed_html_page = BeautifulSoup(scraper.get("https://makeup.com.ua/categorys/3/#offset={}".format(36*i)).content,'html.parser')
    for j in parsed_html_page.find_all('li',{'class': 'simple-slider-list__item'}):
        try:
            addParfume(j['data-id'])
        except:
            continue
```
