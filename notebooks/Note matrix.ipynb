```python
import pandas as pd
import pyodbc 
from config import server, database, username, password 
import warnings
warnings.filterwarnings('ignore')
```

```python
def writeToDB():
    cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER='+server+';DATABASE='+database+';UID='+username+';PWD='+ password)
    df = pd.read_sql("select * from dbo.parfumes", cnxn)
    return df
```

```python
df = writeToDB()
```

```python
df
```

```python
parfumeType = []
for i in df['ParfumeType']:
    try:
        for j in i.split(','):
            note = j.strip()
            parfumeType.append(note) if note not in parfumeType else parfumeType
    except:
        continue
parfumeType.sort()
```

# Base notes

```python
base_notes = []
for i in df['BaseNotes']:
    try:
        for j in i.split(','):
            note = j.strip()
            base_notes.append(note) if note not in base_notes else base_notes
    except:
        continue
base_notes.sort()
```

# Heart notes

```python
heart_notes = []
for i in df['HeartNotes']:
    try:
        for j in i.split(','):
            note = j.strip()
            heart_notes.append(note) if note not in heart_notes else heart_notes
    except:
        continue
heart_notes.sort()
```

# Top notes

```python
top_notes = []
for i in df['TopNotes']:
    try:
        for j in i.split(','):
            note = j.strip()
            top_notes.append(note) if note not in top_notes else top_notes
    except:
        continue
top_notes.sort()
```

```python
notes = []
notes = base_notes + heart_notes + top_notes
notes = list(set(notes))
notes.sort()
```
